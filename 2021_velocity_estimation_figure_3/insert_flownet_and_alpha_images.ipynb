# Plot alphas

```python
K = [[615.8287353515625, 0.0, 320.0995178222656],
     [0.0, 615.9053344726562, 245.37171936035156],
     [0.0, 0.0, 1.0]]
fov_x = np.arctan(320./K[0][0])*2
fov_y = np.arctan(240./K[1][1])*2
```

```python
# build arrays for making receptive field filters

shape = [448, 640]

alphas_x = np.linspace(-fov_x/2, fov_x/2, 640)
alphas_y = np.linspace(-fov_y/2, fov_y/2, 448)

alphas_x = np.vstack([alphas_x]*shape[0])
alphas_y = np.vstack([alphas_y]*shape[1]).T

alphas_epi = np.sign(alphas_x)*np.sqrt(alphas_x**2 + alphas_y**2)
betas = np.arctan2(alphas_x, alphas_y)
```

# Plot flows

```python
TAG_CHAR = np.array([202021.25], np.float32)
def writeFlow(filename,uv,v=None):
    """ Write optical flow to file.
    If v is None, uv is assumed to contain both u and v channels,
    stacked in depth.
    Original code by Deqing Sun, adapted from Daniel Scharstein.
    """
    nBands = 2
    if v is None:
        assert(uv.ndim == 3)
        assert(uv.shape[2] == 2)
        u = uv[:,:,0]
        v = uv[:,:,1]
    else:
        u = uv
    assert(u.shape == v.shape)
    height,width = u.shape
    f = open(filename,'wb')
    # write the header
    f.write(TAG_CHAR)
    np.array(width).astype(np.int32).tofile(f)
    np.array(height).astype(np.int32).tofile(f)
    # arrange into matrix form
    tmp = np.zeros((height, width*nBands))
    tmp[:,np.arange(width)*2] = u
    tmp[:,np.arange(width)*2 + 1] = v
    tmp.astype(np.float32).tofile(f)
    f.close()
```

# V over d

```python
flownet_data = np.load( os.path.join(data_directory, 'flo_705_741.npy'))
```

```python
cs = np.dstack(((np.cos(alphas_epi)*np.sin(alphas_epi))*np.sign(np.sin(alphas_epi))/np.cos(alphas_epi), 
                (np.cos(alphas_epi)*np.sin(alphas_epi))*np.sign(np.sin(alphas_epi))/np.cos(alphas_epi)))

def plot_v_over_d(flow, ax):
    flow[:,:,1] = flow[:,:,1]*-1
    v_over_d = flow / cs 
    
    mags = np.dstack((np.linalg.norm((v_over_d), axis=2),np.linalg.norm((v_over_d), axis=2)))
    
    ax.imshow( np.linalg.norm((v_over_d), axis=2), vmin=0, vmax=1.2, origin='lower', cmap='YlGn',
           extent=[np.min(alphas_x), np.max(alphas_x), np.min(alphas_y), np.max(alphas_y)])

    v_over_d_direction = (v_over_d ) #/ mags) 
    res = 50
    i = int(res/2.)
    X = alphas_x[i::res, i::res]
    Y = alphas_y[i::res, i::res]
    U, V = v_over_d_direction[i::res,i::res,0], v_over_d_direction[i::res,i::res,1]
    q = ax.quiver(X, Y, U, V)#, scale=100, linewidths=0.4, width=0.0055)
    
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlim(alphas_x[0][0], alphas_x[-1][-1])
    ax.set_ylim(alphas_y[0][0], alphas_y[-1][-1])
    fifi.mpl_functions.adjust_spines(ax, [])
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)
f = copy.copy(flownet_data)
plot_v_over_d(f[29, ::-1, :, :], ax)
```

```python
layout = fifi.svg_to_axes.FigureLayout('optic_flow_summary_figure.svg', autogenlayers=True, make_mplfigures=True, hide_layers=[])

ax = layout.axes[('v_over_d', 'v_over_d_1')]
#flow_file_name = '../images/flownet_raw/flo-raw-522.npy'
flow = flownet_data[29, ::-1, :, :]
plot_v_over_d(flow, ax)


ax = layout.axes[('v_over_d', 'v_over_d_2')]
#flow_file_name = '../images/flownet_raw/flo-raw-527.npy'
flow = flownet_data[35, ::-1, :, :]
plot_v_over_d(flow, ax)



layout.append_figure_to_layer(layout.figures['v_over_d'], 'v_over_d', cleartarget=True)

layout.write_svg('optic_flow_summary_figure.svg')
```
